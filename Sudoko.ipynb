```python
#Converting Given input sudoko series inro numpy array for further processing
def preprocessing(lst):
    print('Input string : ', lst)
    
    #list with 9 elements each type = string
    l =[]
    for i in (0, 9, 18, 27, 36,45,54,63, 72):
        l.append(lst[i: i+9])
    #print(l)
    
    
    #Created list with all chars as integer
    final = []
    for line in l:
        temp = []
        for val in line:
            temp.append(int(val))
        final.append(temp)
    
    #print('Final list is : \n', final)
    
    
    arr =  np.array(final)
    return arr

lst = '004006079000000602056092300078061030509000406020540890007410920105000000840600100'   
#lst = '060900400070300006409050103090731000350209018000685020207090601600008040001002030'
output_arr =  preprocessing(lst)
print('Output array : \n',output_arr)
```

```python

def block_creation(i, j, arr):
    block = []
    #block 0,1,2
    if  0<=j & j<3 and  0<=i & i<3:
        for i in range(3):
            for j in range(3):
                block.append(arr[i][j])
    
    
    if  3<=j & j<6 and  0<=i & i<3:
        for i in range(3):
            for j in range(3, 6):
                block.append(arr[i][j])
                
                
    if  6<=j & j<9 and  0<=i & i<3:
        for i in range(3):
            for j in range(6, 9):
                block.append(arr[i][j])
                
    
    #block 3,4,5
    if  0<=j & j<3 and  3<=i & i<6:
        for i in range(3, 6):
            for j in range(3):
                block.append(arr[i][j])
    
    
    if  3<=j & j<6 and  3<=i & i<6:
        for i in range(3,6):
            for j in range(3, 6):
                block.append(arr[i][j])
                
                
    if  6<=j & j<9 and  3<=i & i<6:
        for i in range(3,6): 
            for j in range(6, 9):
                block.append(arr[i][j])
                
    #block 6,7,8
    if  0<=j & j<3 and  6<=i & i<9:
        for i in range(6,9):
            for j in range(3):
                block.append(arr[i][j])
    
    
    if  3<=j & j<6 and  6<=i & i<9:
        for i in range(6, 9):
            for j in range(3, 6):
                block.append(arr[i][j])
                
                
    if  6<=j & j<9 and  6<=i & i<9:
        for i in range(6,9):
            for j in range(6, 9):
                block.append(arr[i][j])
    
    return block

print(block_creation(0,2, arr))
```

```python
#Creating list of elements not avaliable for given block 

def block_values(i, j, arr):
    
    temp = list(filter(lambda x: x!= 0, block_creation(i,j, arr)))
    temp = list(filter(lambda x: x not in temp, digits))
    return temp
    
block_values(0, 3, arr) 
```

```python
def zeros(arr):
    count = 0
    for i in arr:
        for j in i:
            if j==0:
                count +=1
    return count
```

```python
#Creating list of elements not avaliable for given column 
def column_values(j, arr):
    temp=[]
    
    for i in range(9):
        temp.append(arr[i][j])
    
    temp1 = list(filter(lambda x: x!= 0, temp))
    
    temp2 = list(filter(lambda x: x not in temp1, digits))
    return temp2
print(column_values(0,arr))
```

```python
modified_arr = copy.deepcopy(arr)


#Solving sudoko using row, colum, block non elements
def suduko_manipulation(arr, temp_arr):
    
    #Check for if any update is done or not
    flag = False
    
    #To keep the track of number of updates with given logic
    count = 0
    print('Array to update : ')
    print(temp_arr)
    
    
    for i in range(9):
        for j in range(9):
            
            #Value to update to slove sudoko
            if arr[i][j] ==0:
                print('Position to work on: ', i, j)
        
                row_items = row_values(i, temp_arr)
                column_items = column_values(j, temp_arr)
                block_items = block_values(i, j, temp_arr)
        
                print('row : {0} \ncolumnn: {1} \nblock : {2}'.format(row_items, column_items, block_items))
        
        
                updated_row =  list(filter(lambda x:  x in block_items, row_items))
                #print(updated_row)
        
                updated_column =  list(filter(lambda x:  x in block_items, column_items))
                #print(updated_column)
        
                common_items = list(filter(lambda x:  x in updated_row, updated_column))
                #print(common_items)
        
                key = str(i)+str(j)
                dict[key]= common_items
        
                
                #Possible options to update is one 
                if len(common_items) == 1:
                    print('Before updation : \n', temp_arr)
                    print('Updating..... Value to update: ',common_items[0] )
                    temp_arr[i][j]= common_items[0]
                    count +=1
                    print('After updation : \n', temp_arr)
                
                #More options to update
                else:
                    print('possible options are more.. NO UPDATION  \n')
                    
            
            #No need to update on given position as value is available
            else:
                print('No need to work on position ({},{}) value is {}'.format(i, j, temp_arr[i][j]))
            
            print()
            
    print('Total updates done in given iterations: ', count)
    if count>0:
        flag = True
        
    print('Updated value of flag is : ', flag)
    
    
    return (temp_arr, flag)
```

```python
def Value_not_avalable(array):
    not_avaialble = list(filter(lambda x: x not in array, digits))
    if len(not_avaialble)==1:
        
        print('Value to be added : ', not_avaialble[0])
        return not_avaialble[0]
    
    elif len(not_avaialble) ==2:
        print('Values to be added : ', not_avaialble)
        return not_avaialble
Value_not_avalable([9, 2, 0, 0, 6, 7, 8, 4, 3])
```

```python
#Function for updating sudoko for single value for ROWS
def rowwise_update(arr):
    flag = False
    rows=  []
    
    #Created all rows in list
    for i in range(9):
        temp = []
        for j in range(9):
            temp.append(arr[i][j])
        #print(temp, i)
        rows.append(temp)
    
    
    #Finding total zereos and updating rows with single zero
    for row in rows:
    
        count =0
        for value in row:
        
            if value == 0:
                count+=1
        print('Zeros in {} : {} '.format(row, count))
    
        if count == 1:
            flag = True
            VALUE_TO_ADD = Value_not_avalable(row)
            print('Value to be added : ', VALUE_TO_ADD)
            
            for value in row:
                if value == 0:
                    print(row.index(value))
                    row[row.index(value)]= VALUE_TO_ADD
                    break
    return (np.array(rows), flag)
                
rowwise_arr, row_flag = rowwise_update(modified_array)
```

```python
#Function for updating sudoko for single value for Columns
def columnwise_update(arr):
    columns, column_flag = rowwise_update(np.transpose(arr))
    columns =  np.transpose(np.array(columns))
    return (columns, column_flag)

columnwise_arr, column_flag = columnwise_update(modified_array)
```
